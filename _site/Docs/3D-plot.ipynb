```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

plt.rc('font', size=15)
```

# First of all, try your best to interpret your data into 2-D figures rather than 3-D. 3-D figures are not friendly for paper readers.

```python
x, y = np.meshgrid(np.arange(-5, 6, 1), np.arange(-5, 6, 1))
z = x**2 + y**2
```

### Now `z` is a 2-D array.

```python
fig, ax = plt.subplots(figsize=(5, 4.5))
plt.pcolormesh(x, y, z, cmap='jet')
plt.xlabel('x')
plt.ylabel('y')
cb = plt.colorbar()
cb.ax.set_ylabel('z')
cb.ax.tick_params(direction='in')
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(x, y, z)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
sfc = ax.plot_surface(x, y, z, cmap='jet', linewidth=0, antialiased=False)
fig.colorbar(sfc)
```
